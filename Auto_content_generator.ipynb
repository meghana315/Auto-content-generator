```python
from transformers import pipeline

# Initialize the text generation pipeline with GPT-2 model (you can use other models like GPT-3 if available)
generator = pipeline("text-generation", model="gpt2")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

def generate_adventure_story(prompt, max_length=800, num_variations=1):
    # Use a larger GPT-2 model for better storytelling
    model_name = "gpt2-xl"  # Try "gpt2-large" or "gpt2-medium" if RAM is limited
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # Ensure the tokenizer has a padding token
    tokenizer.pad_token = tokenizer.eos_token

    # Encode the prompt
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text with improved settings
    generated_content = []
    for _ in range(num_variations):
        outputs = model.generate(
            inputs,
            max_length=max_length,
            num_return_sequences=1,
            do_sample=True,
            top_p=0.92,
            temperature=0.8,
            no_repeat_ngram_size=4,
            repetition_penalty=1.3,
            pad_token_id=tokenizer.pad_token_id,
            eos_token_id=tokenizer.eos_token_id
        )

        # Decode the generated text and remove the original prompt if repeated
        content = tokenizer.decode(outputs[0], skip_special_tokens=True).replace(prompt, "").strip()
        generated_content.append(content)

    return generated_content

# **New Prompt - Starts With a Scene Instead of Instructions**
user_prompt = """The jungle was alive with the sounds of insects and distant howls. Alex wiped the sweat from his brow, his machete slicing through the thick vines blocking his path.

He had spent years chasing rumors of a lost city hidden deep within the Amazon. Now, as he stepped into a clearing, his heart pounded. Before him stood towering stone pillars covered in ancient glyphs—proof that the legends were true.

But before he could take another step, the ground beneath him trembled. A deep grinding sound echoed through the trees. Something was moving beneath the earth.

Suddenly, the ground gave way, and Alex plunged into darkness..."""

# Generate the story
story = generate_adventure_story(user_prompt, max_length=800, num_variations=1)

# Print the result
print("Generated Adventure Story:\n")
print(story[0])
```
